## CRAWLING -  BeautifulSoup & Selenium

```python
def turn_to_num(string):
    
    num = re.sub("[^0-9]", "", string)
    
    if len(num) < 3:
        size = 0
    else:
        size = int(num)
    
    return size
```

```python
def extract_apartment_with_soup(url,n_pages,min_num,max_num,break_num): #soup_Neot_Ashelim
    
    time.sleep(10)
    

    chrome_options = Options()
    chrome_options.add_argument("--kiosk")
    driver = webdriver.Chrome(executable_path = PATH,chrome_options=chrome_options)
    driver.get(url)
    time.sleep(10)
    
    list_adrass = [] 
    list_date = []
    list_squre = []
    list_price_for_meter = []
    list_num_of_rooms = []
    list_flor = []
    list_year_build = []
    
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser') # html.parser # lxml
    first_source = soup
    
    size = 0
    while size < n_pages:
        i = 0
        for aper in soup.find_all('div', class_= 'css-1l81fir e12olmym10'):
            #Before and after have Null
            if (i > min_num and i < max_num):
                list_adrass.append(aper.find('div',class_='css-6aofk1 ex3b8kr9').text)
                list_date.append(aper.find('div',class_='css-rxx06w ex3b8kr4').text)
                temp = aper.find('div',class_='css-1yfqtqv ex3b8kr0').text
                temp = turn_to_num(temp)
                list_price_for_meter.append(temp)
                #list_price_for_meter.append(aper.find('div',class_='css-1yfqtqv ex3b8kr0').text)
                
            i = i + 1
            if i == break_num:# After this info don׳t relvent
                break
            
            
            j = 0
            for r in aper.find_all('div',class_ = 'table-cell css-gfoud7 e12olmym4'):
                if j == 0:
                    list_squre.append(r.find('div',class_='css-14f4wf6 e3vrfmg6').text) # טוב
                if j == 1:
                    list_num_of_rooms.append(r.find('div',class_='css-14f4wf6 e3vrfmg6').text)
                if j == 2:
                    list_flor.append(r.find('div',class_='css-14f4wf6 e3vrfmg6').text)
                if j == 3:
                    list_year_build.append(r.find('div',class_='css-14f4wf6 e3vrfmg6').text)
                    
                j = j+1
                ## END for num 2
                
            #END for num 1 
            
        size = size +1
        dan = driver.find_element_by_class_name("css-1smt1r7")
        dan.find_element_by_class_name("css-mpwnkp").click() # go to next page 
        
        time.sleep(10)
        
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser') # source for new page
        
    ### END WHILE
    
   
    df = pd.DataFrame({"adrass":list_adrass,"date":list_date,"squre":list_squre,"price for meter":list_price_for_meter,"num of rooms":list_num_of_rooms,"flor":list_flor,"year build":list_year_build})
    
    return df
```

```python
def info_about_neighborhood(url):
    
    time.sleep(10)
    
    chrome_options = Options()
    chrome_options.add_argument("--kiosk")
    driver = webdriver.Chrome(executable_path = PATH,chrome_options=chrome_options)
    driver.get(url)
    time.sleep(10)

    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser') # html.parser # lxml
    
    i = 0
    for aper in soup.find_all('div', class_= 'css-1rsaddu erkaxs93'):
        if i == 1:
            Average_rent = aper.text
            Average_rent = turn_to_num(Average_rent)
        if i == 2 :
            School_grade = aper.text
            
        if i == 5 :
            Socio_economic = aper.text  
            
        i = i + 1

    i = 0
    for aper in soup.find_all('div', class_= 'css-l6tjk3 e3vrfmg5'):
        if i == 0:
            Cleanliness_Maintenance = aper.text  
            
        if i == 2:
            Kindergarten = aper.text     
            
        if i == 3:
            Parking = aper.text 
            
        if i == 4:
            Gardens = aper.text##
            
        if i == 5:
            Recreation_and_leisure = aper.text #
            
        if i == 6:
            Public_Transport = aper.text
            
        if i == 7:
            Feeling_confident = aper.text
        
        if i == 9:
            Shopping = aper.text #
            
        if i == 11:
            Pedestrian_comfort = aper.text #
            
        i = i+1
        
    
    df_info_neighborhood = pd.DataFrame({"Average_rent":Average_rent,"School_grade":School_grade,"Socio_economic":Socio_economic,"Cleanliness_Maintenance":Cleanliness_Maintenance,"Kindergarten":Kindergarten,"Parking":Parking,"Public_Transport":Public_Transport,"Feeling_confident":Feeling_confident,
                                        "Gardens":Gardens,"Recreation_and_leisure":Recreation_and_leisure,"Shopping":Shopping,"Pedestrian_comfort":Pedestrian_comfort}, index=[0])  
    
    
    return df_info_neighborhood
```
